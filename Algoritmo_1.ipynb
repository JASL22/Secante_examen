 - JHONN SAETEROS



# CÓDIGO 

```python
X_n = [1/2**i for i in range(100)] # X_n = x_1, x_2, hasta el infinito
N_x = len(X_n)

SUM = 0
tolerancia = 0.1  # Error permitido ya que 10^-1
limite_suma = 2  # Ya que se va a proximando a 2 al sumar consecutivamente
numero_minimo_para_error = 0

for i in range(N_x):
    SUM = SUM + X_n[i]
    error = limite_suma - SUM
    if error < tolerancia:
        numero_minimo_para_error = i + 1  # +1 porque los índices inician en 0
        break

# Resultados
print("Número mínimo de términos necesarios:", numero_minimo_para_error)
print("La suma parcial es:", SUM)
print("El error absoluto es:", error)
```
